```python
#Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
#A3. 
# Arithmetic Error includes:
# ZeroDivisionError
# OverFlowError
# FloatingPointError

# Zero Division Error: The error which occurs when any number is divided by zero.For Ex:
try:
    a=1/0
except Exception as e:
    print(e.__class__)
```

```python
# OverFlow Error: This error occurs when the output of a clculation exceeds the maximum limit of a data type.For Ex:
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except Exception as e:
    print(e)
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
# A4. LookupError class is the base class for all exceptions that are raised when an index or a key is not
#     found for a sequence or a dictionary respectively.You can use LookupError exception class to handle 
#     both IndexError and KeyError exception classes.

# IndexError: The error which occurs when the given index is out of range in a given list, tuple or string.
x = [1, 2, 3, 4]
try:
    print(x[10])
except Exception as e:
    print(e.__class__)
```

```python
# Key Error: The error which occurs when the given key is not present in the given dictinary.
dict={'name':'pwskills',
      'course':'Data Science',
      'price':'3500'
}
try:
    print(dict['teacher_name'])
except Exception as e:
    print(e.__class__)
```

```python
# Q6. List down some best practices for exception handling in python.
# A6. Some of the best practices for exception handling ar listed below:
```

```python
# 1.Use always a specific exception
try :
    10/0
except ZeroDivisionError as e :
    print(e)
```

```python
# 2.Print always a proper message 
try :
    10/0
except ZeroDivisionError as e :
    print("i am trying to handle a zerodivision error"  , e)
```

```python
# 3.Always try to log your error
import logging
logging.basicConfig(filename = "error.log" , level = logging.ERROR)
try :
    10/0
except ZeroDivisionError as e :
    logging.error("i am trying to handle a zerodivision error {} ".format(e) )
```

```python
# 4.Document all the error 
# 5.Cleanup all the resources
try :
    with open("test.txt" , 'w') as f :
        f.write("this is my data to file " )
except FileNotFoundError as e : 
    logging.error("i am handling file not found  {} ".format(e) )
finally :
    f.close()
```
